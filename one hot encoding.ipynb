### One Hot Encoding
One hot encoding is one method of converting data to prepare it for an algorithm and get a better prediction. With one-hot, we convert each categorical value into a new categorical column and assign a binary value of 1 or 0 to those columns. Each integer value is represented as a binary vector.

```python
df=pd.read_csv("D:/csv files/cars.csv")
df.tail()
```

## One Hot Encoding using sklearn

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=3)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
x_train_new=encoder.fit_transform(x_train[['fuel','owner']])
x_test_new=encoder.fit_transform(x_test[['fuel','owner']])
```
